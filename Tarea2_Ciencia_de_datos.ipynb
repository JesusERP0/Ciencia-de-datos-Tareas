Algo que si me interesa saber por curioidad es, ¿cuantos tipos(por asi decirlo) de animes hay?. Hacemos un histograma obteniendo: 

```python
plt.hist(data.Type)
plt.show()
```

Como podemos ver y resulta natural es que hay mas series de anime tipo "TV" que los demas tipos, algo que se me hace muy impresionante es la cantidad de peliculas que hay, que se encuentran por encima de las OVA´S.

Me encontre esto por ahi y lo aplique para ver lo de los animes mas vistos por genero.

```python
genders = []
genders_count = {}

for gender in data['Tags'].fillna('').tolist():
    gender_array = [x for x in str(gender).split(',') if x != '' and x != 'nan']
    genders.extend(gender_array)

genders = np.unique(genders).tolist()

for gender in genders:
    genders_count.update({
        gender: len(data[data['Tags'].fillna('').str.contains(gender)])
    })

plt.figure(figsize=(30,60))
plt.pie(list(genders_count.values()), labels=list(genders_count.keys()))
plt.show()
```

Podeos ver que esta todo encimado jaja, al parecer ahora si me aparecen todos los animes. Viendo los animes que hay para cada genero tenemos:

```python
genders_sort = sorted(genders_count.items(),  reverse=False)

for i in genders_sort:
    l_gen = list(i)
    print( l_gen)
    
```

El genero con mas animes es el de accion con 3761 animes.
